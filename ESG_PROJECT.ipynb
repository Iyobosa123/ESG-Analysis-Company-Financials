To know if the percentage of missing values is insignificant(small) so I can drop it

```python
total_rows=len(df)
print(total_rows)
missing_growth=df["GrowthRate"].isna().sum()
print(missing_growth)
percentage_missing_growth=(missing_growth/total_rows)
print(percentage_missing_growth)
```

```python
df=df.dropna(subset=["GrowthRate"])
```

To confirm whether a decrease in carbon emmission, water usage and energy consumption would result to an increase in  ESG_environmental as expected, I did a correlation analysis below. The negative values show that there is an inverse relationship, as it should be. Though the relationship is weak

```python
correlation = df['CarbonEmissions'].corr(df['ESG_Environmental'])
print(f'Correlation between Carbon Emissions and ESG Environmental: {correlation:.2f}')
```

```python
correlation = df['WaterUsage'].corr(df['ESG_Environmental'])
print(f'Correlation between Water Usage and ESG Environmental: {correlation:.2f}')
```

```python
correlation = df['EnergyConsumption'].corr(df['ESG_Environmental'])
print(f'Correlation between Energy Consumption and ESG Environmental: {correlation:.2f}')
```

To confirm if the sum of values of carbon emission, water usage and energy consumption columns in my dataset would equate to the values in the ESG_environmental column as expected

```python
df['ESG_sum'] = (df['ESG_Environmental'] + df['ESG_Social'] + df['ESG_Governance'])/3
print(df[['ESG_sum', 'ESG_Overall']])
```

The financial metrics are of different scales; Revenue is in the thousands, ProfitMargin is around 10 and GrowthRate is around 5. In other to compare them across different ESG categories in one plot, I have to normalize them so the big values (Revenue) would not dominate and hide the others.

```python
import matplotlib.pyplot as plt
df['ESG_Category'] = pd.qcut(df['ESG_Overall'], q=3, labels=['Low', 'Medium', 'High'])
grouped = df.groupby('ESG_Category', observed=True)[['ProfitMargin', 'Revenue', 'GrowthRate']].mean()
```

```python
grouped = df.groupby('ESG_Category', observed=True)[['ProfitMargin', 'Revenue', 'GrowthRate']].mean()
print(grouped)
```

```python
import matplotlib.pyplot as plt
grouped = df.groupby('ESG_Category', observed=True)[['ProfitMargin', 'Revenue', 'GrowthRate']].mean()
normalized = grouped / grouped.max()

normalized.T.plot(kind='bar', figsize=(10, 6), colormap='Set2', title='Normalized Financial Metrics by ESG Category')
plt.ylabel('Relative Scale (0-1)')
plt.xlabel('Financial Metric')
plt.grid(True)
plt.legend(title='ESG Category')
plt.tight_layout()
plt.xticks(rotation=0)
plt.show()
plt.savefig("financial.png", dpi=300, bbox_inches='tight')

```

To determine if the size of a company affects its involvement in ESG

```python
# Create size category: Small, Medium, Large
df['SizeCategory'] = pd.qcut(df['MarketCap'], q=3, labels=['Small', 'Medium', 'Large'])
```

```python
# Group by company size and calculate average ESG scores
grouped_esg = df.groupby('SizeCategory')[['ESG_Overall', 'ESG_Environmental', 'ESG_Social', 'ESG_Governance']].mean()
print(grouped_esg)
```

```python
# Plot ESG scores by company size
grouped_esg.plot(kind='bar', figsize=(10, 6), colormap='Set2', title='Average ESG Scores by Company Size')
plt.ylabel('Average ESG Score')
plt.xlabel('Company Size')
plt.grid(True)
plt.legend(title='ESG Component')
plt.tight_layout()
plt.xticks(rotation=0)
plt.show()
plt.savefig("size.png", dpi=300, bbox_inches='tight')

```



To determine the level of ESG performance across Regions



```python
# Group by Region and calculate average ESG_overall
region_sorted = df.groupby('Region')['ESG_Overall'].mean().sort_values()

# Display the sorted result
print(region_sorted)
```

```python
import matplotlib.pyplot as plt
region_sorted.plot(kind='barh', figsize=(8, 5), color='skyblue', title='Average ESG Overall Score by Region (Ascending)')
plt.xlabel('ESG Overall Score')
plt.ylabel('Region')
plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig("region.png", dpi=300, bbox_inches='tight')

```
